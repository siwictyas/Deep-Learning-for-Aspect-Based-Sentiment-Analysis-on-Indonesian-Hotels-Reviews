```python
model_harga = load_model('/content/drive/My Drive/Thesis/Code/FixModel/gru_sentiment_harga_9048.h5')
model_hotel = load_model('/content/drive/My Drive/Thesis/Code/FixModel/gru_sentiment_hotel_9145.h5')
model_kamar = load_model('/content/drive/My Drive/Thesis/Code/FixModel/clstm_sentiment_kamar_9114.h5')
model_lokasi = load_model('/content/drive/My Drive/Thesis/Code/FixModel/cnn_sentiment_lokasi_9565.h5')
model_pelayanan = load_model('/content/drive/My Drive/Thesis/Code/FixModel/clstm_sentiment_pelayanan_9084.h5')
model_restoran = load_model('/content/drive/My Drive/Thesis/Code/FixModel/clstm_sentiment_restoran_9179.h5')
```

```python
text_harga = harga["text"]
text_hotel = hotel["text"]
text_kamar = kamar["text"]
text_lokasi = lokasi["text"]
text_pelayanan = pelayanan["text"]
text_restoran = restoran["text"]

tokenizer_harga = Tokenizer()
tokenizer_hotel = Tokenizer()
tokenizer_kamar = Tokenizer()
tokenizer_lokasi = Tokenizer()
tokenizer_pelayanan = Tokenizer()
tokenizer_restoran = Tokenizer()

tokenizer_harga.fit_on_texts(text_harga.values)
tokenizer_hotel.fit_on_texts(text_hotel.values)
tokenizer_kamar.fit_on_texts(text_kamar.values)
tokenizer_lokasi.fit_on_texts(text_lokasi.values)
tokenizer_pelayanan.fit_on_texts(text_pelayanan.values)
tokenizer_restoran.fit_on_texts(text_restoran.values)

max_seq_len_harga = max([len(i.split()) for i in text_harga])
max_seq_len_hotel = max([len(i.split()) for i in text_hotel])
max_seq_len_kamar = max([len(i.split()) for i in text_kamar])
max_seq_len_lokasi = max([len(i.split()) for i in text_lokasi])
max_seq_len_pelayanan = max([len(i.split()) for i in text_pelayanan])
max_seq_len_restoran = max([len(i.split()) for i in text_restoran])

list_test_harga = text_harga.tolist()
list_test_hotel = text_hotel.tolist()
list_test_kamar = text_kamar.tolist()
list_test_lokasi = text_lokasi.tolist()
list_test_pelayanan = text_pelayanan.tolist()
list_test_restoran = text_restoran.tolist()

seq_harga = tokenizer_harga.texts_to_sequences(list_test_harga)
seq_hotel = tokenizer_hotel.texts_to_sequences(list_test_hotel)
seq_kamar = tokenizer_kamar.texts_to_sequences(list_test_kamar)
seq_lokasi = tokenizer_lokasi.texts_to_sequences(list_test_lokasi)
seq_pelayanan = tokenizer_pelayanan.texts_to_sequences(list_test_pelayanan)
seq_restoran = tokenizer_restoran.texts_to_sequences(list_test_restoran)

padded_harga = pad_sequences(seq_harga, maxlen=max_seq_len_harga)
padded_hotel = pad_sequences(seq_hotel, maxlen=max_seq_len_hotel)
padded_kamar = pad_sequences(seq_kamar, maxlen=max_seq_len_kamar)
padded_lokasi = pad_sequences(seq_lokasi, maxlen=max_seq_len_lokasi)
padded_pelayanan = pad_sequences(seq_pelayanan, maxlen=max_seq_len_pelayanan)
padded_restoran = pad_sequences(seq_restoran, maxlen=max_seq_len_restoran)

pred_harga = model_harga.predict(padded_harga)
pred_hotel = model_hotel.predict(padded_hotel)
pred_kamar = model_kamar.predict(padded_kamar)
pred_lokasi = model_lokasi.predict(padded_lokasi)
pred_pelayanan = model_pelayanan.predict(padded_pelayanan)
pred_restoran = model_restoran.predict(padded_restoran)
```
